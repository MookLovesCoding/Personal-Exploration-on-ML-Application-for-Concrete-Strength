```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Mike\Documents\concrete strength ML\data\raw\Concrete_Data.csv")
```

```python
# renaming columns
df.columns = [
    "cement",
    "slag",
    "fly_ash",
    "water",
    "superplasticizer",
    "coarse_agg",
    "fine_agg",
    "age",
    "strength"
]
df.head()
df = df.dropna()
df.to_csv(r"C:\Users\Mike\Documents\concrete strength ML\data\processed\concrete_data_cleaned.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 8)

df.hist(bins=30)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# correlation matrix
corr = df.corr(numeric_only=True)

# draw
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation matrix of numeric features")
plt.show()
```

just observing the heatmap we can expect stuff with more cement, older age, and more superplasticizer to be stronger. Not too sure what else I should be looking for in the histograms and heatmap though.
